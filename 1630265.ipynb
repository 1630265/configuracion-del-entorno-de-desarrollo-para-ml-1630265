```python
import random
import matplotlib.pyplot as pl
import numpy as np

matriz = []
for i in range(1000):
    matriz.append([])
    for j in range(2):
        matriz[i].append(random.randint(-100,100))
        
x = np.array(matriz)
y = np.array(matriz)
pl.scatter(x[:,0],y[:,1])
pl.title('Matriz puntos')
pl.xlabel('$matriz[i]$')
pl.ylabel('$matriz[j]$')
pl.show()
print(matriz)
```
